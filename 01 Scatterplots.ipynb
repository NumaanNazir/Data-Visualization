```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOAD A DATA SET OF CHURN MODELLING
x=pd.read_csv("C:\\Users\\nqnqa\\Downloads\\Churn_Modelling.csv")
```

```python
x.columns
```

```python
#SCATTER PLOT WITH PANDAS
x.plot.scatter(x="Age" ,y="Balance")
```

```python
#SCATTER PLOT WITH SEABORN
sns.scatterplot(data=x,x="Age" ,y="Balance")
```
