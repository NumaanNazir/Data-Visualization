```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOAD A DATA SET OF CHURN MODELLING
x=pd.read_csv("C:\\Users\\nqnqa\\Downloads\\Churn_Modelling.csv")
```

```python
x.columns
```

```python
#GROUPBY For plotting Bar plot's

a=x.groupby(["Gender"])[["Balance","CreditScore"]].mean().reset_index()

print(a)
```

```python
# Barplot with Pandas 

pandas=a.plot.bar(x="Gender"	,y="CreditScore")

print(pandas)
```

```python
#Barplot with  Matplotlib

mat=plt.bar(a["Gender"]	,a["CreditScore"])

print(mat)
```

```python
#Barplot with  Seaborn

sea=sns.barplot(data=a,x="Gender",y="CreditScore")

print(sea)
```

```python
#HORIZONTAL BARPLOT  >>  

# Horizontal Barplot with Pandas 

pandas=a.plot.barh(x="Gender"	,y="CreditScore")

print(pandas)
```

```python
#Barplot with  Matplotlib

mat=plt.barh(a["Gender"]	,a["CreditScore"])

print(mat)
```
